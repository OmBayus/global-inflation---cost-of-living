## Linear Regression - Turkey Inflation

```python
data = pd.read_csv("google_trends_inflation_rate.csv")
data = data.drop(['Indicator Name',
       'Indicator Code', 'Alpha-2 Country Code', 'Lang1', 'ISO 639-1 Code',
       'HL', 'inflation_translation'],axis=1)
```

```python
turkey = data[data["Country Name"] == "Turkey"].iloc[0:1,3:63]
turkey
```

```python
from sklearn.model_selection import train_test_split

x = turkey.columns.astype(int).to_numpy()
y = turkey.astype(float).iloc[0,:].to_numpy()

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.30)
```

> Shuffle dataset and Split %70 of the dataset for training and %30 for testing

```python
from sklearn.linear_model import LinearRegression

# create model
lr = LinearRegression()

# reshape the data to 2d array
x_train = x_train.reshape(-1,1)

# train the model
lr.fit(x_train,y_train)
```

> Train the model with train data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# reshape the data to 2d array
x_test = x_test.reshape(-1, 1)

# predict the test data
pr = lr.predict(x_test)

print("Mean Squared Error: ",mean_squared_error(y_test, pr))
print("Mean Absolute Error: ",mean_absolute_error(y_test, pr))
```

```python
# fit a polynomial of degree 1
line = np.polyfit(x, y, 1)
# create the polynom
poly = np.poly1d(line)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=x, y=y,
                    mode='markers',
                    name='inflation rate'))
fig.add_trace(go.Scatter(x=x, y=poly(x),
                    mode='lines',
                    name='polynom'))

fig.show()
```

### Predict Turkey inflation rate in 2023

```python
year_2023 = np.array([2023])
year_2023 = year_2023.reshape(-1, 1)
predict_2023_inflation = lr.predict(year_2023)
predict_2023_inflation[0]
```
