## Linear Regression - Annual inflation mean of all countries

```python
from sklearn.model_selection import train_test_split

x = years.astype(int).to_numpy()
y = mean_years.astype(float).to_numpy()

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.30)
```

> Shuffle dataset and Split %70 of the dataset for training and %30 for testing

```python
from sklearn.linear_model import LinearRegression

# create model
lr = LinearRegression()

# reshape the data to 2d array
x_train = x_train.reshape(-1,1)

# train the model
lr.fit(x_train,y_train)
```



> Train the model with train data



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# reshape the data to 2d array
x_test = x_test.reshape(-1, 1)

# predict the test data
pr = lr.predict(x_test)

print("Mean Squared Error: ",mean_squared_error(y_test, pr))
print("Mean Absolute Error: ",mean_absolute_error(y_test, pr))
```

```python
# fit a polynomial of degree 1
line = np.polyfit(x, y, 1)
# create the polynom
poly = np.poly1d(line)
# show instances

import plotly.graph_objects as go

fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=x, y=y,
                    mode='markers',
                    name='inflation rate'))
fig.add_trace(go.Scatter(x=x, y=poly(x),
                    mode='lines',
                    name='polynom'))

fig.show()
```

### R² SCORE

```python
from sklearn.metrics import r2_score

r2 = r2_score(y,poly(x))
r2
```

## Predict world inflation mean in 2023

```python
year_2023 = np.array([2023])
year_2023 = year_2023.reshape(-1, 1)
predict_2023_inflation = lr.predict(year_2023)
predict_2023_inflation[0]
```

# World Bank - Inflation, GDP deflator 1961 - 2021

https://data.worldbank.org/indicator/NY.GDP.DEFL.KD.ZG

```python
# Get Data
world_bank_data = pd.read_csv("world_bank_all_countries.csv")
display(world_bank_data)
```

### Delete unused columns

```python
world_bank_data = world_bank_data.drop(['Indicator Name',
       'Indicator Code'],axis=1)
display(world_bank_data)
```

```python
world_bank_data.info()
```

```python
import plotly.graph_objects as go

world_bank_mean_years = world_bank_data.iloc[:,3:].mean()
world_bank_years = world_bank_data.columns[3:]

google_trends_mean_years = data.iloc[:,3:63].mean()

fig = go.Figure(layout_title_text="According to World Bank and Google Trends, Annual inflation mean of all countries")
fig.add_trace(go.Bar(
    x=world_bank_years,
    y=world_bank_mean_years,
    name='World Bank',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=world_bank_years,
    y=google_trends_mean_years,
    name='Google Trends',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

## Polynomial Regression - Annual inflation mean of all countries

### 2nd degree polynomial

```python
from sklearn.model_selection import train_test_split

x = world_bank_years.astype(int).to_numpy()
y = world_bank_mean_years.astype(float).to_numpy()

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.30)
```

> Shuffle dataset and Split %70 of the dataset for training and %30 for testing

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)

x_train = x_train.reshape(-1, 1)

x_poly = pf.fit_transform(x_train)

lr2 = LinearRegression()
lr2.fit(x_poly,y_train)
```

> Train the model with train data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# reshape the data to 2d array
x_test = x_test.reshape(-1, 1)

# predict the test data
pr = lr2.predict(pf.fit_transform(x_test))

print("Mean Squared Error: ",mean_squared_error(y_test, pr))
print("Mean Absolute Error: ",mean_absolute_error(y_test, pr))
```

### Predict world inflation mean in 2023

```python
year_2023 = np.array([2023])
year_2023 = year_2023.reshape(-1, 1)
predict_2023_inflation = lr2.predict(pf.fit_transform(year_2023))
predict_2023_inflation[0]
```

## Support Vector Regression - Annual inflation mean of all countries

```python
from sklearn.preprocessing import StandardScaler

x = world_bank_years.astype(int).to_numpy()
y = world_bank_mean_years.astype(float).to_numpy()

x = x.reshape(-1,1)
y = y.reshape(-1,1)

sc1 = StandardScaler()

X_scale = sc1.fit_transform(x)

sc2 = StandardScaler()
Y_scale = sc2.fit_transform(y)
```

> Train the model

```python
from sklearn.svm import SVR

svr_reg = SVR(kernel="rbf")
svr_reg.fit(X_scale,Y_scale)

import plotly.graph_objects as go

fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=X_scale[:,0], y=Y_scale[:,0],
                    mode='markers',
                    name='inflation rate'))
fig.add_trace(go.Scatter(x=X_scale[:,0], y=svr_reg.predict(X_scale),
                    mode='lines',
                    name='polynom'))

fig.show()
```

### Predict world inflation mean in 2023

```python
year_2023 = np.array([2023])
year_2023 = year_2023.reshape(-1, 1)
predict_2023_inflation = svr_reg.predict(year_2023)
predict_2023_inflation[0]
```
