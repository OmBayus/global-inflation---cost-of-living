```python
import matplotlib.pyplot as plt
import pandas as pd

f = pd.read_csv('world bank global inflation rate.csv')
world = f.groupby('Country Name').get_group('World')
attribute_names = list(world.to_dict().keys())[4:]

x_axis = attribute_names
y_axis = []

for name in attribute_names:
  y_axis.append(dict(world[name])[259])


fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(5)

plt.plot(x_axis, y_axis, color='red')
plt.title('Global Inflation Rate (World Bank)')
plt.ylabel('Inflate Rate in %')
plt.xticks(rotation = 45)
plt.show()
```

```python
import graphviz
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

X = data.drop(columns=['Country','Cost of Living Index']) 
y = data['Cost of Living Index'].apply(lambda y_data: 1 if y_data>=50 else 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
fig = plt.figure(figsize=(10,10))
model =Pipeline([ 
    ('tree', DecisionTreeClassifier())
])
model.fit(X_train, y_train)

# Visualize
annotations = plot_tree(model['tree'], filled=True) 
```

```python
X=data.iloc[:,2:].values
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,5): 
  kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10,random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,5),wcss)
plt.title('Cost of Living Index 2022')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init ='k-means++', max_iter=300, n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X)
```

```python
scatter_list = [
    {"kmean":"0","color":"red"},
    {"kmean":"1","color":"blue"},
    {"kmean":"2","color":"green"},
    {"kmean":"3","color":"cyan"},
    {"kmean":"4","color":"magenta"},
]

for i in scatter_list:
  plt.scatter(X[y_kmeans==int(i["kmean"]), 0], X[y_kmeans==int(i["kmean"]), 1], s=100, c=i["color"], label ='Cluster '+i["kmean"])

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')
plt.title('Cost of Living Index 2022')
plt.xlabel('Index Rates')
plt.ylabel('Scores')
plt.show()
```
